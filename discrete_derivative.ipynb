```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return -2*(x**2)+50*x

x = np.linspace(start=-20, stop=50, num=1000)
y = f(x)

plt.plot(x,y)
plt.grid();
```
