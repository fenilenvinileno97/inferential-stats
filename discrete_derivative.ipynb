By a discrete differential approach, when we retrieve the limit of a function when h tends towards zero, it is possible to get a good approximation of its derivative. For instance, if we derive the original function `f(x)`: 

$$f(x) = -2x^2 + 50x$$
$$\frac{df}{dx} = -4x + 50$$

And when we evaluate `f'(x)` in `x = 0`, we should retrieve the coordinates `(x, f'(x)) = (0, 50)`

### Derivative of a function by discrete differential approach.

```python
def dx(f):
    h = 1e-6
    return ((f(x+h)-f(x))/h) # This approach turns out the discrete approximation of a derivative function.
```

```python
y = square_f(x)

x_point_derivate = 0
y_point_derivate = 50

# plt.plot(x,y, 'b')
plt.plot(x,dx(square_f), 'r')
plt.grid()
plt.xlim(-5,5)
plt.ylim(20,80)
plt.plot(x_point_derivate, y_point_derivate, marker="o", color="black")
plt.hlines(y_point_derivate, xmin=-10, xmax=x_point_derivate, colors="green", linestyles="--")
plt.vlines(x_point_derivate, ymin=0, ymax=y_point_derivate, colors="green", linestyles="--")
plt.text(x_point_derivate, y_point_derivate, f'({x_point_derivate}, {y_point_derivate})', fontsize=12, ha='left', va='bottom')
plt.title(r"Derivative: f'($x) = -4$x + 50")
plt.tight_layout();
```

It can be observed that coordinates `(0, 50)` are found in our derivative's plot conducted by discrete approach. We could conclude that our approach is quite close.

### Derivative of Heaviside function.

$$
H(x) = \begin{cases}
0, & \text{if} \ \ x < 0 \\
1, & \text{if} \ \ x \geq 0
\end{cases}
$$

```python
def heaviside_f(x):
    Y = np.zeros(len(x))
    for idx, x in enumerate(x):
        if x>=0:
            Y[idx] = 1
    return Y
```

```python
z = np.linspace(start = -100, stop = 100, num = 1000)
plt.plot(z, heaviside_f(z), '-', color='black')
plt.plot(z, dx(heaviside_f), 'y--', alpha=0.75)
plt.ylim(-0.1, 1.1)
plt.xlim(z.min(), z.max())
plt.plot(0, 1, 'ro')
plt.plot(0, 0, 'ro', fillstyle='none')
plt.grid();
```

### Derivative of hyperbolic tangent

```python
def hyperbolic_tan(x):
    return (np.exp(2*x) - 1) / (np.exp(2*x) + 1)
```

```python
plt.plot(z, hyperbolic_tan(z))
plt.plot(z, dx(hyperbolic_tan))
plt.plot(z, (1-(hyperbolic_tan(z))**2))
plt.xlim(-4, 4)
plt.grid();
```

### Derivative of sigmoidal function.

```python
def sigmoidal_f(x):
    return 1/(1 + np.exp(-x))
```

```python
plt.plot(z, sigmoidal_f(z))
plt.plot(z, dx(sigmoidal_f))
plt.plot(z, (sigmoidal_f(z)*(1-sigmoidal_f(z))))
plt.xlim(-6,6)
plt.grid();
```

### Derivative of ReLU function.

$$
R(x) = \begin{cases}
0, & \text{if} \ \ x < 0 \\
x, & \text{if} \ \ x \geq 0
\end{cases}
$$

$$
R'(x) = H(x) = \begin{cases}
0, & \text{if} \ \ x < 0 \\
1, & \text{if} \ \ x \geq 0
\end{cases}
$$

```python
def reLU(x):
    return np.maximum(x, 0)
```

```python
plt.plot(z, reLU(z))
plt.plot(z, heaviside_f(z))
plt.ylim(-1, 10)
plt.xlim(-10, 10)
plt.grid();
```
